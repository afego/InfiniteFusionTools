```python
import requests
from bs4 import BeautifulSoup
```

```python
req_url = 'https://infinite-fusion-calculator.com/pokedex/'
# success code - 200
r = requests.get(req_url)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
# De acordo com uma anÃ¡lise rÃ¡pida da saÃ­da do BeautifulSoup
# Todos os pokemons estao no elemento 'a' com a classe abaixo
dex = soup.find_all("a", class_="grid grid-cols-2 md:grid-cols-6 gap-1 md:gap-2 bg-gray-400/10 rounded-sm my-1 p-1")
print(type(dex))
for d in dex:
    # o atributo 'href' do elemento 'a' descreve o nome do pokemon, enquanto o 'id' sua posiÃ§Ã£o na dex
    print(d["href"], d["id"])
```

#### Criando lista de pokÃ©mons com seus nomes e tipos para se tornar um DataFrame

```python
pokedex_list = []
id = 1
for pokemon in dex:

    species = pokemon.find_all("span", class_="pl-2 self-center")[0].text
    
    types = pokemon.find_all("div", class_="flex gap-1 self-center")[0].find_all("img")
    type1 = types[0]["src"][12:-4]
    if len(types)> 1:
        type2 = types[1]["src"][12:-4]
    else:
        type2 = type1

    attrs = []
    pkm_attrs = pokemon.find_all("div", class_="flex col-span-2 text-sm text-gray-600 dark:text-gray-300")[0]
    i = 1
    for att in pkm_attrs.find_all("span"):
        if i%2 == 0:
            attrs.append(att.text)
        i += 1
    
    curr_pokemon = [id, species,type1, type2]
    curr_pokemon += attrs
    pokedex_list.append(curr_pokemon)
    id += 1
```

```python
import pandas
```

```python
df = pandas.DataFrame(pokedex_list, columns = ["DexId","Species","Type1","Type2","HP","ATK","DEF","SATK","SDEF","SPE"])
df.head(6)
```
