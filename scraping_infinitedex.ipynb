Obtendo informações da pokedex de Infinite Fusion a partir do site não oficial "Infinite Dex"
- Quais são os Pokémons inclusos.
- Quais são os tipos, em ordem de cada Pokémon. Visto que alguns estão com a ordem trocada.

Atualizado em Agosto/2024, version 6.2.3

```python
import requests
from bs4 import BeautifulSoup
```

```python
req_url = 'https://infinitefusiondex.com/pokedex'
# success code - 200
r = requests.get(req_url)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
# De acordo com uma análise rápida da saída do BeautifulSoup
# Todos os pokemons estao no elemento 'a' com a classe abaixo
dex = soup.find_all("a", class_="col info-row p-0 m-0 section highlight d-flex justify-content-evenly")
print(type(dex))
for d in dex:
    # o atributo 'href' do elemento 'a' descreve o nome do pokemon
    print(d["href"])
```

```python
sample_pkm = dex[0]
print(type(sample_pkm))
print(sample_pkm)
print(sample_pkm["href"])
```

##### Procurando os tipos

```python
pkm_types = sample_pkm.find_all("div", class_="d-flex justify-content-center col-md-4 col-4")[0]
print(pkm_types)
```

```python
# Só há um elemento div com a classe especificada, então acesso logo o primeiro resultado Tag do ResultSet
types = pkm_types.find_all("img")
print(types)
```

```python
# pkm_type_1 = types[0]["src"]
# pkm_type_2 = types[1]["src"]
# print(pkm_type_1, pkm_type_2)
type_list = []
for type in types:
    type_list.append(type["alt"])
print(type_list)
```

#### Criando lista de pokémons com seus nomes e tipos para se tornar um DataFrame

```python
pokedex_list = []
for pokemon in dex:

    dex_id = pokemon.findChild().text
    
    species = pokemon.find_all("div", class_="text-center text-nowrap col-md-4 col-4")[0].text
    
    types = pokemon.find_all("div", class_="d-flex justify-content-center col-md-4 col-4")[0].find_all("img")
    type1 = types[0]["alt"]
    if len(types)> 1:
        type2 = types[1]["alt"]
    else:
        type2 = type1
    
    curr_pokemon = [dex_id, species,type1, type2]
    pokedex_list.append(curr_pokemon)
```

```python
import pandas
```

```python
df = pandas.DataFrame(pokedex_list, columns = ["DexId","Name","Type1","Type2"])
df.head(6)
```

```python
df.to_csv("infinitefusiondex.csv", index=False)
```
