```python
import pandas as pd
```

```python
pokedex = pd.read_csv("data/CompleteInfiniteFusionDex.csv")
pokedex.head(6)
```

### Search by type

```python
def search_type(type,type_slot=0,verbose=True):
    '''
    type_slot:
        0 = Primary or secondary type
        1 = Type1, Primary type
        2 = Type2, Secondary type
        3 = monotype
    '''
    if type_slot == 3:
        result = pokedex.loc[(pokedex["Type1"] == type) & (pokedex["Type2"] == type)]
    elif type_slot == 0:
        result = pokedex.loc[(pokedex["Type1"] == type) | (pokedex["Type2"] == type)]
    else:
        result = pokedex.loc[(pokedex[f"Type{type_slot}"] == type)]
    if verbose:
        print(result["Species"].values)
    return result
```

```python
requested_type = 'fire'
type_slot = 2

res_type_1 = search_type(requested_type, type_slot)
```

```python
requested_type = 'normal'
type_slot = 3

res_type_2 = search_type(requested_type, type_slot)
```

### Search by ability

```python
def search_ability(ability,verbose=True):
    result = pokedex.loc[(pokedex["ability_1"].str.lower() == ability) | (pokedex["ability_2"].str.lower() == ability) | (pokedex['ability_hidden'].str.lower() == ability)]
    if verbose:
        print(result["Species"].values)
    return result
```

```python
requested_ability = "healer"
res_ability_1 = search_ability(requested_ability)
```

```python
requested_ability = "chlorophyll"
res_ability_2 = search_ability(requested_ability)
```

### Sorting results

```python
column_sort = ['speed','sp_attack']
res_type_1[res_type_1.total >= 450].sort_values(by=column_sort,ascending=False)
```

```python
column_sort = 'speed'
res_ability_2[res_ability_2.total >= 450].sort_values(by=[column_sort],ascending=False)
```
