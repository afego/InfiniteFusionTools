```python
import pandas as pd
```

```python
pokedex = pd.read_csv("data/CompleteInfiniteFusionDex_v6.2.3.csv")
pokedex.head(6)
```

```python
def sort_df(df, columns:list, bst:int=0, ascending:bool=False):
    columns = [x.lower() for x in columns]
    return df[df.bst >= bst].sort_values(by=columns,ascending=ascending)
```

### Search by type

```python
def search_type(type,type_slot=0,verbose=True):
    '''
    type_slot:
        0 = Primary or secondary type
        1 = type_1, Primary type
        2 = type_2, Secondary type
        3 = monotype
    '''
    if type_slot == 3:
        result = pokedex.loc[(pokedex["type_1"] == type) & (pokedex["type_2"] == type)]
    elif type_slot == 0:
        result = pokedex.loc[(pokedex["type_1"] == type) | (pokedex["type_2"] == type)]
    else:
        result = pokedex.loc[(pokedex[f"type_{type_slot}"] == type)]
    if verbose:
        print(result["name"].values)
    return result
```

```python
requested_type = 'rock'
type_slot = 0

res_type_1 = search_type(requested_type, type_slot)
```

```python
requested_type = 'ground'
type_slot = 0

res_type_2 = search_type(requested_type, type_slot)
```

### Search by ability

```python
def search_ability(ability,verbose=True):
    result = pokedex.loc[(pokedex["ability_1"].str.lower() == ability) | (pokedex["ability_2"].str.lower() == ability) | (pokedex['ability_hidden'].str.lower() == ability)]
    if verbose:
        print(result["name"].values)
    return result
```

```python
requested_ability = "healer"
res_ability_1 = search_ability(requested_ability)
```

```python
requested_ability = "chlorophyll"
res_ability_2 = search_ability(requested_ability)
```

### Sorting results

```python
column_sort = ['DexId']
min_bst = 0
sort_df(
    res_type_1,
    column_sort,
    bst=min_bst,
    )
```

```python
column_sort = ['DexId']
min_bst = 0
sort_df(
    res_type_2,
    column_sort,
    bst=min_bst,
    )
```
