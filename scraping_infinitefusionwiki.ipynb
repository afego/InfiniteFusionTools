Obtendo informações da pokedex de Infinite Fusion a partir do site não oficial "Infinite Fusion Fandom"
- Quais são os Pokémons inclusos.
- Os nomes dos Pokémons com suas devidas formas
- Seus tipos em ordem no jogo e.g. Empoleon é Steel/Water em vez de Water/Steel

Atualizado em Agosto/2024, versão 6.2.3

```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
req_url = 'https://infinitefusion.fandom.com/wiki/Pokédex'
# success code - 200
r = requests.get(req_url)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
# De acordo com uma análise rápida da saída do BeautifulSoup
# Todos os pokemons estao no elemento 'a' com a classe abaixo
dex = soup.find_all("tr")
pokedex = []
id = 1
for d in dex[2:]:
    # o atributo 'text' do elemento 'a' descreve o nome do pokemon
    # print(len(d))
    if len(d) >= 12:
        tds = d.find_all("td")
        # print(tds)
        curr_pkm = []
        form = []
        for td in tds:
            form.extend(td.find_all("small"))
            curr_pkm.extend(td.find_all("a",class_="external text"))
        # tds = d.find_all(re.compile("https://infinitefusiondex.com/details/\d+"))
        print(id, len(curr_pkm))
        # print(form)
        pkm = [id,curr_pkm[0].text] # Infinite Fusion Dex Id e Nome
        if form:
            pkm.append(form[0].text)
        else:
            pkm.append("Base")
        if len(curr_pkm) == 3: # Tornando o tipo 2 igual ao tipo 1, caso nao tenha tipo secundario
            pkm.extend([curr_pkm[1].text,curr_pkm[2].text])
        else:
            pkm.extend([curr_pkm[1].text,curr_pkm[1].text])
        pokedex.append(pkm)
        id+=1
    print("===============")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(pokedex, columns=["DexId","Name","Form","Type 1","Type 2"])
df.head(6)
```

```python
df.to_csv("data/infinitefusionwiki.csv", index=False)
```
